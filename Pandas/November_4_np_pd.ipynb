Factorial of a number without a function

```python
n = int (input ("Enter a number: "))
factorial = 1
if n >= 1:
    for i in range (1, n+1):
        factorial=factorial*i
print("Factorial of the given number is: ", factorial)
```

Factorial function

```python
def fact(num):
  fact=1
  if (num==0) or (num==1):
     return 1
  else:
     for i  in range(num,1,-1):
      fact*=i
  return fact

fact(4)
```

Try using Recurssion Sub-String Search

```python
def count_occurences(s, sub):
    if len(s) == 0:
        return 0
    else:
        ind = s.find(sub)

        if ind>=0:
            return 1+count_occurences(s[ind+1:], sub)
        else:
            return 0
count_occurences("banana","ana")
```

```python
def count_substring(s, sub):
    if len(sub)>len(s):
      return 0
    return s.startswith(sub) + count_substring(s[1:],sub)
count_substring("malayalam","la")
```
