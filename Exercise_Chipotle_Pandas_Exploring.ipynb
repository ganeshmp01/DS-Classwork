### Step 9. Which was the most ordered item?

```python
c=chipo.groupby('item_name').sum()
c=c.sort_values(['quantity'], ascending=False)
c.head(1)
```

### Step 10. How many items were ordered?

```python
chipo.quantity.sum()
```

### Step 11. What was the most ordered item in the choice_description column?

```python
a=chipo.groupby('choice_description').sum()
a=a.sort_values(['quantity'], ascending=False)
a.head(1)
```

### Step 12. How many items were orderd in total?

```python
chipo.quantity.sum()
```

### Step 13. Turn the item price into a float

```python
chipo.item_price
```

```python
chipo.item_price=chipo.item_price.apply(lambda x: x.replace(' ','$'))
chipo.item_price=chipo.item_price.astype(object)
chipo.item_price
```

### Step 14. How much was the revenue for the period in the dataset?

```python
chipo.head()
```

```python
chipo['revenue'] = chipo['quantity']*chipo['item_price']
chipo['revenue'].sum()
```

```python
chipo.head()
```

### Step 15. How many orders were made in the period?

```python
chipo['order_id'].value_counts().count()
```

### Step 16. What is the average amount per order?

```python
chipo.groupby('order_id')['revenue'].agg('mean')
```
