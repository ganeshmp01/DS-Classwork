```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fifa_file=pd.read_csv("data-for-datavis/fifa.csv",index_col="Date",parse_dates=True)
fifa_file
```

```python
sns.lineplot(data=fifa_file)
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=fifa_file)
```

```python
spotify_data=pd.read_csv("data-for-datavis/spotify.csv",index_col="Date",parse_dates=True)

spotify_data
```

```python
plt.figure(figsize=(20,10))
plt.title("Populate songs streamed between 2017-18")
plt.xlabel("Dates from 2017-18")
sns.lineplot(data = spotify_data)
```

```python
#Figure size setting
plt.figure(figsize = (20,10))
plt.title("Populat songs streamed between 2017-18")
plt.xlabel("Dates from 2017-18")
sns.lineplot(data=spotify_data['Despacito'],label = 'Desp')
sns.lineplot(data=spotify_data['Shape of You'],label='SOY')
```
